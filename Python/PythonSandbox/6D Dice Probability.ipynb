```python
import pandas as pd
import random
import seaborn as sns
```

```python
results_dict = {
    i:{j:0 for j in range(1, 13)} for i in range(1, 7)
}
results_dict[1]
```

```python
def roll():
    return random.randint(1, 6)
```

```python
def up_to_12():
    i = 0
    sub_sum = 0
    while sub_sum < 13:
#         print(f'Sum: {sub_sum}, i: {i}')
        rolled = roll()
        sub_sum += rolled
        i += 1
        results_dict[rolled][i] += 1
#     print(f'Sum: {sub_sum}, i: {i}')
#     print('---------------------------')
    return [sub_sum, i]
```

```python
def play(n):
    for i in range(n):
        yield up_to_12()
        
res = play(100000)
```

```python
df = pd.DataFrame(res)
```

```python
df.columns = ['Sum', 'Rolls']
```

```python
sns.set(rc={'figure.figsize':(10,6)})
df['Sum'].hist(bins=len(df['Sum'].unique()))
```
